```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf
import os,time
```

```python
data = tf.keras.datasets.mnist
```

```python
(X_train,y_train),(X_test,y_test) = data.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
plt.imshow(X_train[54000])
plt.show()
```

```python
# here we will take the 5000 rows as validation part and the remaining part as a training 
X_valid , X_train = X_train[:5000]/255. , X_train[5000:]/255.
y_valid , y_train = y_train[:5000], y_train[5000:]

#  We are dividing with 255 so that we can get the range of 0-1
#  its like the normalize the pixel value
#  over here we are using backpropagation so that the conversion of global minima is very fast
```

```python
X_test = X_test / 255.
```

```python
# flatten means convert 2d or 3d data into 1d
Layers = [tf.keras.layers.Flatten(input_shape=[28,28],name="input_layer"),
         tf.keras.layers.Dense(20,activation="relu",name="hidden_layer1"),
         tf.keras.layers.Dense(15,activation="relu",name="hidden_layer3"),
         tf.keras.layers.Dense(10,activation="softmax",name="output_layer"),]
```

```python
model = tf.keras.models.Sequential(Layers)
model.summary()
```

```python
loss = "sparse_categorical_crossentropy"
optimizer = "ADAM"
metrics = ["accuracy"]
```

```python
model.compile(loss=loss,optimizer = optimizer,metrics=metrics)
```

```python
hidden1 = model.layers[1]
```

```python
weights,bias = hidden1.get_weights()
```

```python
# by default the batch size is 32
# so over here we have 5000/32
EPOCHS = 20
validation_set  = (X_valid,y_valid)
history = model.fit(X_train,y_train,epochs=EPOCHS,validation_data= validation_set)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
X = df1[['Caption', 'Followers', 'Hashtags']]
y_like = df1['Likes']
y_time = df1['Time since posted']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_like_train, Y_like_test, Y_time_train, Y_time_test = train_test_split(X,y_like,y_time,random_state=21,test_size=0.2)
```

```python
label_encoder = LabelEncoder()
```

```python
X_train_encoded = X_train.copy()
X_test_encoded = X_test.copy()


for col in X_train.columns:
    X_train_encoded[col] = label_encoder.fit_transform(X_train[col])
    

for col in X_test.columns:
    X_test_encoded[col] = label_encoder.fit_transform(X_test[col])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train_encoded, Y_like_train)
likes_predictions = lr.predict(X_test_encoded)

mse = mean_squared_error(Y_like_test, likes_predictions)

print("Mean Squared Error (MSE):", mse)
```

```python
lr.fit(X_train_encoded, Y_time_train)
posted_time_prediction = lr.predict(X_test_encoded)

mse = mean_squared_error(Y_time_test, posted_time_prediction)
mse
```
